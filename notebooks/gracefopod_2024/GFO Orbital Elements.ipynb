```python
#!/usr/bin/env python3

import fnmatch
from ftplib import FTP
import os

ftp = FTP('isdcftp.gfz-potsdam.de')

#if no username and password is required.
ftp.login()

# # if you have to change directory on FTP server.
ftp.cwd('/grace-fo/Level-1B/JPL/INSTRUMENT/RL04/2022/')

# Get all files
files = ftp.nlst()


path_save = "/data/SatDragModelValidation/data/inputs/raw_inputdata/data_GRACEFO/Level1B_isdcftp/"

# Download files
for i,file in enumerate(files):

    if os.path.exists(path_save+file):
        print(f" {i} --- already downloaded, {file}")
        
    elif fnmatch.fnmatch(file, '*noLRI.tgz'):  #To download specific files.
        print(f" {i} Downloading..." + file)

        try:
            ftp.retrbinary("RETR " + file ,open(path_save + file, 'wb').write)

        except EOFError:    # To avoid EOF errors.
            pass
    else:
        pass

ftp.close()
```

```python
def RV_to_OE(R_vec, V_vec, mu):
    import numpy as np
    R_vec = np.array(R_vec) #km
    V_vec = np.array(V_vec )
#     r = np.sqrt(R_vec[0]**2 + R_vec[1]**2 + R_vec[2]**2) 
#     v = np.sqrt(V_vec[0]**2 + V_vec[1]**2 + V_vec[2]**2) 
    r = np.linalg.norm(R_vec)
    v = np.linalg.norm(V_vec)



    h_vec = np.array(np.cross(R_vec,V_vec))
    h = np.linalg.norm(h_vec)
    
    p = (h**2/mu)

    n_vec = np.cross(np.array([0,0,1]),h_vec)
    n = np.linalg.norm(n_vec)


    energy = ((v**2)/2)-(mu/r)

    e_vec = (((v**2)-(mu/r))*R_vec - (np.dot(R_vec,V_vec))*V_vec)/mu

    a = -1*(mu/(2*energy))
    e = np.linalg.norm(e_vec)
    i = np.arccos(h_vec[2] / h)*(180/np.pi)

    RAAN = np.arccos(n_vec[0]/n)*(180/np.pi)
    if n_vec[1]<0:
#         RAAN = -1*RAAN
        RAAN = 360-RAAN #-1*arg_peri

    arg_peri = np.arccos(np.dot( n_vec, e_vec )/ (n * e) )*(180/np.pi)
    if e_vec[2]<0:
        arg_peri = 360-arg_peri #-1*arg_peri
#         arg_peri = -1*arg_peri

    true_anom = np.arccos(np.dot(e_vec,R_vec)/(e*r))*(180/np.pi)
    if np.dot(R_vec,V_vec) < 0:
        true_anom = -1*true_anom
    
    vr= (mu/h)*(e*np.sin(true_anom*(np.pi/180)))

    v_theta = (mu/h)*(1 + e*np.cos(true_anom*(np.pi/180)))

    r_vec_rot = np.array([r, 0,0])
    v_vec_rot = np.array([vr, v_theta,0])
    
    Period = (2*np.pi)*np.sqrt(a**3/mu)
    

#     print('inc =',i)
#     print('a =',a)
#     print('raan =', RAAN)
#     print('ecc =',e)
#     print('arg_periapsis =',arg_peri)
   
    
    return(i, a, RAAN, e, arg_peri)

```

```python
import pandas as pd
import os
import numpy as np

dates = pd.date_range(start='01/01/2022', end='01/31/2022')
dirdata = "/data/SatDragModelValidation/data/inputs/raw_inputdata/data_GRACEFO/Level1B_isdcftp/"

fns = dates
# Allocate variables
inc      = np.zeros(np.size(fns))
a        = np.zeros(np.size(fns))
RAAN     = np.zeros(np.size(fns)) 
ecc      = np.zeros(np.size(fns))
arg_peri = np.zeros(np.size(fns))


# f = open('OE_dailyaverages_gfo', "w")
# f.write("\n")
# f.close()

# ### Write to file
# with open('OE_dailyaverages_gfo', 'r+') as file:

for ifn,val in enumerate(dates):
    print(f"{val.year}-{val.month:02d}-{val.day:02d}")
    if f"{val.year}-{val.month:02d}-{val.day:02d}" == '2022-03-31' or \
       f"{val.year}-{val.month:02d}-{val.day:02d}" == '2022-06-22' or \
       f"{val.year}-{val.month:02d}-{val.day:02d}" == '2022-09-16' :

        print(f"skipping {val.year}-{val.month:02d}-{val.day:02d}") 
        inc[ifn]      = np.nan
        a[ifn]        = np.nan
        RAAN[ifn]     = np.nan
        ecc[ifn]      = np.nan
        arg_peri[ifn] = np.nan
        continue




    ### Identify the files for that day
    __tarfile = f"gracefo_1B_{val.year}-{val.month:02d}-{val.day:02d}_RL04.ascii.noLRI.tgz"
#     __gnifile = f"GNI1B_{val.year}-{val.month:02d}-{val.day:02d}_C_04.txt"
    __gnvfile = f"GNV1B_{val.year}-{val.month:02d}-{val.day:02d}_C_04.txt"

    
    if os.path.exists(f"{dirdata}pod_files/{__gnvfile}"):
        print(f" --- already in folder, {__gnvfile}")  
        pass

    else:
        print(f" Untaring and copying..." + __gnvfile)

        ### untar/gunzip the the file of interest and copy to directory.
        os.system(f"tar -xvzf {dirdata}{__tarfile} {__gnvfile} --directory {dirdata}pod_files/")
#         os.system(f"mv {__gnifile} {dirdata}pod_files/")
        os.system(f" mv {__gnvfile} {dirdata}pod_files/")


    ## Load data for this day --------------------------------------------------------------------
    dirpath     = f"{dirdata}pod_files/"
#     filename = f"GNI1B_{val.year}-{val.month:02d}-{val.day:02d}_C_04.txt"   # Use Earth-Centered Inertial coordinate frame
    filename = f"GNV1B_{val.year}-{val.month:02d}-{val.day:02d}_C_04.txt"   # Use Earth-Centered Earth fixed coordinate frame
    datapath     = dirpath + filename
    headers = [
        'gps_time',   #  second,  Continuous seconds past 01-Jan-2000 11:59:47 UTC
        'GRACEFO_id', #           C = GRACE-FO 1 (GRACE C),        D = GRACE-FO 2 (GRACE D)
        'coord_ref',  #           E = Earth-Centered, Earth-Fixed, I = Earth-Centered Inertial 
        'xpos',       #  m,       Position, X value
        'ypos',       #  m,       Position, Y value         
        'zpos',       #  m,       Position, Z value         
        'xpos_err',   #  m,       Formal error of xpos       
        'ypos_err',   #  m,       Formal error of ypos        
        'zpos_err',   #  m,       Formal error of zpos        
        'xvel',       #  m/s,     Velocity along x-axis        
        'yvel',       #  m/s,     Velocity along y-axis         
        'zvel',       #  m/s,     Velocity along z-axis         
        'xvel_err',   #  m/s,     Formal error of xvel
        'yvel_err',   #  m/s,     Formal error of yvel
        'zvel_err',   #  m/s,     Formal error of zvel
        'qualflg',    #         
                ]
    bigdf = pd.read_csv(datapath, 
            skiprows = 148, 
            sep = '\s+',
            names = headers)
    # #Convert date seconds since epoch in GPS to Date UTC
    date = pd.to_datetime(bigdf['gps_time'], unit='s',
                   origin=pd.Timestamp('2000-01-01 11:59:47'))#- pd.to_timedelta(18,'s')

    bigdf.insert(0, 'Date', date)
    del bigdf['xpos_err'], bigdf['ypos_err'], bigdf['zpos_err']
    del bigdf['xvel_err'], bigdf['yvel_err'], bigdf['zvel_err']
    del bigdf['qualflg']
    del bigdf['GRACEFO_id']
    del bigdf['coord_ref']
    del bigdf['gps_time']
    ##  ------------------------------------------------------------------------------------------


    ### Gather R and V to calc the orbital elements for this day
    temp_inc      = np.zeros(np.size(bigdf['Date']))
    temp_a        = np.zeros(np.size(bigdf['Date']))
    temp_RAAN     = np.zeros(np.size(bigdf['Date']))
    temp_ecc      = np.zeros(np.size(bigdf['Date']))
    temp_arg_peri = np.zeros(np.size(bigdf['Date'])) 

    for i,val_i in enumerate(bigdf['Date']):
        r_i = bigdf['xpos'][i]/1000
        r_j = bigdf['ypos'][i]/1000
        r_k = bigdf['zpos'][i]/1000

        v_i = bigdf['xvel'][i]/1000
        v_j = bigdf['yvel'][i]/1000
        v_k = bigdf['zvel'][i]/1000

        r = np.array([r_i,r_j,r_k])
        v = np.array([v_i,v_j,v_k]) 

        ### save the values
        (temp_inc[i],
        temp_a[i],
        temp_RAAN[i],
        temp_ecc[i],
        temp_arg_peri[i]) = RV_to_OE(r, v, mu=398600.4415)

    # Save the daily average
    inc[ifn]      = np.mean(temp_inc)
    a[ifn]        = np.mean(temp_a)
    RAAN[ifn]     = np.mean(temp_RAAN)
    ecc[ifn]      = np.mean(temp_ecc)
    arg_peri[ifn] = np.mean(temp_arg_peri)

#         #### Manually write each row of the data.
#         row =   f"{pd.to_datetime(val).strftime(format='%Y-%m-%d %H:%M:%S')}"\
#             +f"  {inc[ifn]:15.5f}"\
#             +f"  {a[ifn]:15.5f}"\
#             +f"  {RAAN[ifn]:15.5f}"\
#             +f"  {ecc[ifn]:15.5f}"\
#             +f"  {arg_peri[ifn]:15.5f}"\
#             +f"\n"
#         file.write(row)

    print()
```
